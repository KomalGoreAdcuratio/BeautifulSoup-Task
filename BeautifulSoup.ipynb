```python
from pymongo import MongoClient
client=MongoClient('localhost',27017)
db=client['Web_Data']
collection=db['Task1']
```

```python
import requests
from bs4 import BeautifulSoup
url = 'https://news.ycombinator.com/'
r = requests.get(url)
r = r.content
soup = BeautifulSoup(r,'html.parser')
data=soup.find_all('span')

for i in data:
    #print(i.get('class',None))
    if i.get('class',None)==['titleline']:
        mongoData={
            'title'      :    i.find("a").get_text(),
            'link'       :    i.find("a").get("href"),
            'describe'   :    i.find("span").get_text()
        }
        ##print(i.find('a').get('href'),i.find('a').get_text(),i.find('span').get_text(),sep='   ')
        print(f'title      :    {i.find("a").get_text()}')
        
        print(f'link       :    {i.find("a").get("href")}')
        print(f'describe   :    {i.find("span").get_text()}')
        
        print("       ")
        collection.insert_one(mongoData)
        
    
```
